````python
from anthropic import Anthropic

client = Anthropic()
#model = "claude-haiku-4-5"
#model = "claude-sonnet-4-6"
model = "claude-haiku-4-5-20251001"
````

````python
def add_user_messages(messages, text):
    user_message  = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_messages(messages, text):
    assistant_message  = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None, stop_sequences=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if system:
        params["system"] = system
    if stop_sequences:
        params["stop_sequences"] = stop_sequences

    message = client.messages.create(**params)
    return message.content[0].text
````

````python
messages = []

system = """
You are a patient math tutor.
Do not directly answer a student's question.
Guide them to a solution step by step.
"""
add_user_messages(messages, "How do I solve 5x+3=2 for x?")
answer = chat(messages, system=system)

answer
````

````python
messages = []

prompt = """
Generate three different sample AWS CLI commands. Each should be very short.
"""

add_user_messages(messages, prompt)
add_assistant_messages(messages, "Here are all three commands in a single block without any comments:```bash")

text = chat(messages, stop_sequences=["```"])
text.strip()
````
